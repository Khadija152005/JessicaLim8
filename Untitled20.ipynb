```python
import kagglehub

# Download the dataset
path = kagglehub.dataset_download("jonathanoheix/face-expression-recognition-dataset")

print("Dataset downloaded to:", path)
```

```python
import os

# List the contents of the downloaded dataset directory
print("Contents of the downloaded dataset directory:")
for item in os.listdir(path):
    print(item)

# If there are subdirectories, list their contents
for item in os.listdir(path):
    item_path = os.path.join(path, item)
    if os.path.isdir(item_path):
        print(f"\nContents of subdirectory '{item}':")
        for sub_item in os.listdir(item_path):
            print(f"  {sub_item}")
```

```python
import pandas as pd

# Define the paths to the training and validation image directories
train_dir = os.path.join(path, 'images', 'train')
validation_dir = os.path.join(path, 'images', 'validation')

# Initialize lists to store image paths and labels
train_image_paths = []
train_labels = []
validation_image_paths = []
validation_labels = []

# Iterate through the training directory
for label in os.listdir(train_dir):
    label_dir = os.path.join(train_dir, label)
    if os.path.isdir(label_dir):
        for image_file in os.listdir(label_dir):
            image_path = os.path.join(label_dir, image_file)
            train_image_paths.append(image_path)
            train_labels.append(label)

# Iterate through the validation directory
for label in os.listdir(validation_dir):
    label_dir = os.path.join(validation_dir, label)
    if os.path.isdir(label_dir):
        for image_file in os.listdir(label_dir):
            image_path = os.path.join(label_dir, image_file)
            validation_image_paths.append(image_path)
            validation_labels.append(label)

# Create DataFrames to store the information
train_df = pd.DataFrame({'image_path': train_image_paths, 'label': train_labels})
validation_df = pd.DataFrame({'image_path': validation_image_paths, 'label': validation_labels})

# Display the first few rows of the DataFrames and their shapes
print("Training DataFrame:")
display(train_df.head())
print("\nShape of Training DataFrame:", train_df.shape)

print("\nValidation DataFrame:")
display(validation_df.head())
print("\nShape of Validation DataFrame:", validation_df.shape)
```

```python
import cv2
import numpy as np

def preprocess_image(image_path):
    """Loads, resizes, and normalizes a grayscale image."""
    try:
        # Read the image in grayscale
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Warning: Could not read image at {image_path}")
            return None

        # Resize the image to 48x48 pixels
        img = cv2.resize(img, (48, 48))

        # Normalize pixel values to [0, 1]
        img = img / 255.0

        # Convert to NumPy array
        return np.array(img)
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

# Apply the preprocessing function to the image paths
train_df['preprocessed_image'] = train_df['image_path'].apply(preprocess_image)
validation_df['preprocessed_image'] = validation_df['image_path'].apply(preprocess_image)

# Display the first few rows of the updated DataFrames
print("Training DataFrame with preprocessed images:")
display(train_df.head())

print("\nValidation DataFrame with preprocessed images:")
display(validation_df.head())
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

# Extract training data and labels
x_train = np.array(train_df['preprocessed_image'].tolist())
y_train_labels = np.array(train_df['label'].tolist())

# Extract validation data and labels
x_val = np.array(validation_df['preprocessed_image'].tolist())
y_val_labels = np.array(validation_df['label'].tolist())

# Reshape image data to include the channel dimension (grayscale)
# The original shape is (num_samples, height, width)
# The desired shape is (num_samples, height, width, channels) where channels is 1 for grayscale
x_train = x_train.reshape(x_train.shape[0], 48, 48, 1)
x_val = x_val.reshape(x_val.shape[0], 48, 48, 1)

# Encode categorical labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_labels)
y_val_encoded = label_encoder.transform(y_val_labels)

# One-hot encode the labels
y_train = to_categorical(y_train_encoded)
y_val = to_categorical(y_val_encoded)

# Print the shapes of the prepared data and labels
print("Shape of x_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of x_val:", x_val.shape)
print("Shape of y_val:", y_val.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Instantiate a Sequential model
model = Sequential()

# Add convolutional and pooling layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Add a Flatten layer
model.add(Flatten())

# Add dense layers
model.add(Dense(128, activation='relu'))

# Add the output layer
num_classes = 7
model.add(Dense(num_classes, activation='softmax'))

# Display the model summary
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(x_train, y_train,
                    epochs=15,
                    batch_size=64,
                    validation_data=(x_val, y_val))
```

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Paths
train_dir = "/kaggle/input/face-expression-recognition-dataset/images/train"
val_dir = "/kaggle/input/face-expression-recognition-dataset/images/validation"

# Data augmentation & normalization
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

batch_size = 15

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(48, 48),
    color_mode="grayscale",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(48, 48),
    color_mode="grayscale",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False
)

# Model (VGG-like but أخف)
model = Sequential([
    Conv2D(32, (3,3), activation="relu", padding="same", input_shape=(48,48,1)),
    BatchNormalization(),
    Conv2D(32, (3,3), activation="relu", padding="same"),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    Conv2D(64, (3,3), activation="relu", padding="same"),
    BatchNormalization(),
    Conv2D(64, (3,3), activation="relu", padding="same"),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    Conv2D(128, (3,3), activation="relu", padding="same"),
    BatchNormalization(),
    Conv2D(128, (3,3), activation="relu", padding="same"),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    Dropout(0.25),

    Flatten(),
    Dense(256, activation="relu"),
    Dropout(0.5),
    Dense(train_generator.num_classes, activation="softmax")
])

model.summary()

# Compile
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Callbacks
early_stop = EarlyStopping(monitor="val_loss", patience=7, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6)


history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=15,
    callbacks=[early_stop, reduce_lr]
)
```
