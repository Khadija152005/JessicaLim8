```python

df['date'] = pd.to_datetime(df['date'], format="%Y%m%dT%H%M%S")
df['date']
```

```python

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.drop(['date'], axis = 1, inplace = True)
```

```python
plt.figure(figsize = (30, 15))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
df.drop(columns = ['day', 'month', 'year', 'sqft_lot15', 'long', 'zipcode', 'yr_built', 'condition', 'id'], inplace = True)
df
```

```python

df['living_to_lot_ratio'] = df['sqft_living'] / df['sqft_lot']
df.drop(columns = ['sqft_lot', 'sqft_living'], inplace = True)
```

```python
plt.figure(figsize = (30, 15))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.figure(figsize = (20, 5))
sns.boxplot(data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
sqft_living15 = df['sqft_living15']
Q1 = np.percentile(sqft_living15, 25)
Q3 = np.percentile(sqft_living15, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_sqft_living15 = (sqft_living15 >= lower_bound) & (sqft_living15 <= upper_bound)
df = df.loc[filtered_sqft_living15]
```

```python
df['total_sqft'] = df['sqft_above'] + df['sqft_basement']
total_sqft = df['total_sqft']
Q1 = np.percentile(total_sqft, 25)
Q3 = np.percentile(total_sqft, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_total_sqft = (total_sqft >= lower_bound) & (total_sqft <= upper_bound)
df = df.loc[filtered_total_sqft]
```
